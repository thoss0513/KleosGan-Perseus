```python
f = open('parsed_dactylic_hexameter_corpus.json')
data = json.load(f)
data
```

```python
print(data["Homer"].keys())
```

```python
data["Homer"]["Iliad"][1]
```

```python
def add_entry(df, text, author, title, n):
    if n > 0:
        new_row = {'text': text, 'author': author, 'title': title}
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
        
        # Recursively add the next entry
        return add_entry(df, text, author, title, n-1)
    else:
        # Base case: return the DataFrame when n reaches 0
        return df
```

```python
df = pd.DataFrame(columns=['text', 'author', 'title'])
for key in data["Homer"].keys():
    for i in range(len(data["Homer"][key])):
        new_row = {'text': data["Homer"][key][i], 'author': "Homer", 'title': key}
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
```
